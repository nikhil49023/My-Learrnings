```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([5, 7, 8, 10])
y = np.array([12, 18, 16, 20 ])
```

```python
plt.plot(x, y, marker = "o", markersize = 10, markerfacecolor="red", markeredgecolor="green"
         , linestyle = "dotted", linewidth=2, color="green")
plt.grid(axis="x", linewidth=0.5, color="blue") #TO ADD GRID TO  THE GRAPH
 #use (#color code) for colour picked codes
#ADD MARKER TO MAKE PLOTS CUSTOM
```

BAR CHART IS USED TO COMPARE CATEGORIES OF DATA BY REPRESENTING EACH CATEGORY WITH A BAR

```python
cat = np.array(["Vitamins", "Minerals", "Fibre", "Carbs", "Protein"])
vals = np.array([10, 15, 7, 20, 12])
```

```python
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.bar(cat, vals, color = "green")

plt.show()
```

BAR CHART IS A CIRCULAR CHART DIVIDED  INTO SLICES TO SHOW PERCENTAGES OF THE TOTAL. GOOD DOR VISUALIZING DISTRIBUTION AMONG CATEGORIES

```python
plt.pie(vals, labels=cat)
```
