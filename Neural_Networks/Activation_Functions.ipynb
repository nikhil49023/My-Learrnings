```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nnfs
from nnfs.datasets import spiral_data
```

```python
nnfs.init()
X, y = spiral_data(samples=100, classes=3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="brg")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

ReLU is used for Hidden Layers

```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.001 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
        return self.output

class ReLU_Function:
    def forward(self, inputs):
        self.output = np.maximum(0, inputs)
        return self.output

dense1 = Layer_Dense(2, 3)
relu1 = ReLU_Function()

dense1.forward(X)
relu1.forward(dense1.output)

print(relu1.output)
```

```python
That cell defines a tiny “network” and runs a forward pass on the spiral data (passes data through math)

- `Layer_Dense` is a dense (fully connected) layer:
    - `weights` are initialized small (random normal × 0.001) with shape `(n_inputs, n_neurons)`
    - `biases` start as zeros
    - `forward(inputs)` computes `inputs · weights + biases`

- `ReLU_Function` is an activation that zeros out negatives:
    - `forward(inputs)` returns `max(0, inputs)` elementwise

Then it creates:
- `dense1 = Layer_Dense(2, 3)` (takes 2D input, outputs 3 units)
- `relu1 = ReLU_Function()`

And runs:
- `dense1.forward(X)` to get the linear outputs for all input points
- `relu1.forward(dense1.output)` to apply ReLU nonlinearity

Finally it prints the post‑ReLU output, which is the dense layer activations with all negative values set to 0.
```
